```python
#対角線方向に引き伸ばすテスト
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
from IPython.display import display
```

```python
#テスト1
from sklearn.datasets import make_blobs

_,ax = plt.subplots(2,1,figsize=(10,20))



X,_ = make_blobs(random_state=170,n_samples=600)
rng = np.random.RandomState(74)

mglearn.discrete_scatter(X[:,0],X[:,1],ax=ax[0])


transformation = rng.normal(size=(2,2))
X = np.dot(X,transformation)

mglearn.discrete_scatter(X[:,0],X[:,1],ax=ax[1])




plt.xlabel("X")
plt.ylabel("Y")
ax[0].set_title("orinigal function")
ax[1].set_title("transformed function")
```

```python
#テスト2

from sklearn.datasets import make_moons

_,ax = plt.subplots(2,1,figsize=(10,20))



X,_ = make_moons(random_state=170,n_samples=600,noise=0.1)
rng = np.random.RandomState(74)

mglearn.discrete_scatter(X[:,0],X[:,1],ax=ax[0])


transformation = rng.normal(size=(2,2))
X = np.dot(X,transformation)

mglearn.discrete_scatter(X[:,0],X[:,1],ax=ax[1])



plt.xlabel("X")
plt.ylabel("Y")
ax[0].set_title("orinigal function")
ax[1].set_title("transformed function")
```

```python
#テスト3

_,ax = plt.subplots(2,1,figsize=(10,20))

X =np.zeros((100,2))
for i in range(100):
    X[i,0] = np.sin(i)
    X[i,1] = i

mglearn.discrete_scatter(X[:,0],X[:,1],ax=ax[0])


rng = np.random.RandomState(74)
transformation = rng.normal(size=(2,2))
X = np.dot(X,transformation)

mglearn.discrete_scatter(X[:,0],X[:,1],ax=ax[1])



plt.xlabel("X")
plt.ylabel("Y")
ax[0].set_title("orinigal function")
ax[1].set_title("transformed function")
```

```python
#テスト4

_,ax = plt.subplots(2,1,figsize=(10,20))

X =np.zeros((100,2))
for i in range(100):
    X[i,0] = i**(1.2) % 10
    X[i,1] = (i+1) ** (-.05)

mglearn.discrete_scatter(X[:,0],X[:,1],ax=ax[0])


rng = np.random.RandomState(74)
transformation = rng.normal(size=(2,2))
X = np.dot(X,transformation)

mglearn.discrete_scatter(X[:,0],X[:,1],ax=ax[1])



plt.xlabel("X")
plt.ylabel("Y")
ax[0].set_title("orinigal function")
ax[1].set_title("transformed function")
```

```python
#テスト5

_,ax = plt.subplots(2,30,figsize=(240,16))

for a in range(30):
    X =np.zeros((100,2))
    for i in range(100):
        X[i,0] = i
        X[i,1] = i ** ((1.2-0.8)/30*a+0.8)
    mglearn.discrete_scatter(X[:,0],X[:,1],ax=ax[0,a])


    rng = np.random.RandomState(74)
    transformation = rng.normal(size=(2,2))
    X = np.dot(X,transformation)

    mglearn.discrete_scatter(X[:,0],X[:,1],ax=ax[1,a])
    
    
    ax[0,a].set_title("orinigal function\na**{}".format((1.2-0.8)/30*a+0.8))
    ax[1,a].set_title("transformed function")
plt.savefig("対角線に引き延ばす内積計算テスト.png")
```
